```python
#Criando um modelo para prever as doenças em soja utilizando o algoritmo random forest com o objetivo de buscar a melhor performance.
#Os dados são densos e altamente previsíveis
#19 classes
#35 atributos
#69 instâncias
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
dataset=pd.read_csv('/content/soybean.csv')
dataset
```

```python
#Transformando em matriz

previsores= dataset.iloc[:,0:35].values #var.independentes
classe= dataset.iloc[:,35].values #var.dependente
```

```python
#Transformando as var. categóricas em numéricas com o LabelEnconder
#Como são muitas colunas, utilizei um for para cada coluna
labelencoder= LabelEncoder()
for i in range (35):
  previsores[:,i]=labelencoder.fit_transform(previsores[:,i])
```

```python
#Dividindo a base de dados entre treinamento(30%) e teste(70%)
x_train,x_test,y_train,y_test= train_test_split(previsores,classe,test_size=0.3,random_state=0)
x_test
```

```python
#Criação do modelo
floresta= RandomForestClassifier(n_estimators=100)
#treinamento de fato
floresta.fit(x_train,y_train)
```

```python
floresta.estimators_
```

```python
#previsões usando os registros de teste
previsoes= floresta.predict(x_test)
#gerando a matriz de confusão
confusao=confusion_matrix(y_test,previsoes)
```

```python
#taxa de acerto
taxa_acerto=accuracy_score(y_test,previsoes)
taxa_acerto
```
